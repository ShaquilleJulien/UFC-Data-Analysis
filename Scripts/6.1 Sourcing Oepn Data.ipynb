```python
#Check for mixed data
for col in ufc_data.columns.tolist():
    weird = (ufc_data[[col]].map(type) != ufc_data[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (ufc_data[weird]) > 0:
        print (f"Mixed-type data found in column: {col}")
    else: print(f"No mixed-type data in column: {col}")
```

There are 2 columns that have mixed data:
Referee,
B_Stance (Blue fighter fight stance)
R_Stance (Red fighter fight stance)

```python
ufc_data['B_Stance'].value_counts(dropna = False)
```
